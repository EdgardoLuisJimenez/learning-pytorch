```python
import math
import numpy as np
import matplotlib.pyplot as plt

def binomial_coefficient(n, k):
    return (math.factorial(n))/(math.factorial(n-k)* math.factorial(k))

def binomial_distributtion(theta, n, k):
    return binomial_coefficient(n , k) * np.power(theta, k) * np.power((1-theta), n-k)
```

```python
array_1 = np.array([])
array_2 = np.array([])

for i in range(10):
    array_1 = np.append(array_1, binomial_distributtion(0.25, 10, i))
    array_2 = np.append(array_2, binomial_distributtion(0.9, 10, i))
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(8, 4))
ax1.bar(range(10), array_1)
ax2.bar(range(10), array_2)
plt.show()
```

# Gaussian Distribution

```python
import numpy as np
import matplotlib.pyplot as plt

rng = np.random.default_rng(12562)
mu, sigma = 0, np.sqrt(2.25) # Mean and standard deviation
s = rng.normal(mu, sigma, 1000)

fig, ax = plt.subplots()
ax.hist(s, bins=30, color='skyblue', edgecolor='black')
plt.show()
```

# Gradient Descent

```python
import numpy as np
from typing import Callable

def gradient_descent(start: float, gradient: Callable[[float], float],
                     learning_rate: float, max_iter: int, tol: float =0.01):
    x = start
    steps = [start] # History tracking

    for _ in range(max_iter):
        diff = learning_rate*gradient(x)
        if np.abs(diff) < tol:
            break
        x = x - diff
        steps.append(x) # History tracking

    return steps, x
```
